## Question 3

```python
# pick the enrollment measure
enroll_col = 'avg_enrollment'

# total MA enrollment per county-year market
data['total_ma_enroll'] = (
    data.groupby(['ssa', 'year'])[enroll_col]
        .transform('sum')
)

# market share + HHI pieces
data['share'] = data[enroll_col] / data['total_ma_enroll']
data['share_sq'] = data['share'] ** 2

# county-year HHI
hhi_county = (
    data.groupby(['ssa', 'year'])['share_sq']
        .sum()
        .reset_index(name='hhi')
)

# average HHI by year
avg_hhi = (
    hhi_county.groupby('year')['hhi']
        .mean()
        .reset_index()
)

avg_hhi
```

```python
plt.figure()
plt.plot(avg_hhi['year'], avg_hhi['hhi'], marker='o')
plt.xlabel('Year')
plt.ylabel('Average HHI')
plt.title('Average Medicare Advantage HHI, 2014–2019')
plt.show()
```

#### Interpretation: The average HHI for MA markets declined from 2014 to 2019. After a modest decrease from 2014 to 2015, market concentration remained relatively stable through 2017, before falling sharply by 2019. This would suggest Medicare Advantage markets likely became more competitive over time.

## Question 7

```python
df = data[data['year'] == 2018].copy()

# get 2018 HHI (county-year)
hhi_2018 = hhi_county[hhi_county['year'] == 2018][['ssa', 'hhi']].copy()

# merge HHI into df
df = df.merge(hhi_2018, on='ssa', how='left')
```

```python
hhi_33 = hhi_2018['hhi'].quantile(0.33)
hhi_66 = hhi_2018['hhi'].quantile(0.66)

df['market_type'] = 'middle'
df.loc[df['hhi'] <= hhi_33, 'market_type'] = 'competitive'
df.loc[df['hhi'] >= hhi_66, 'market_type'] = 'uncompetitive'
```

```python
df = df[df['market_type'].isin(['competitive', 'uncompetitive'])].copy()
```

```python
df['T'] = (df['market_type'] == 'uncompetitive').astype(int)  # High HHI treated
df['Y'] = df['bid']
```

```python
df['ffs_cost'] = (
    (df['parta_reimb'] + df['partb_reimb']) /
    (df['parta_enroll'] + df['partb_enroll'])
)

ffs = df[['ssa', 'ffs_cost']].drop_duplicates().copy()
ffs['ffs_quartile'] = pd.qcut(ffs['ffs_cost'], 4, labels=[1,2,3,4])
ffs = pd.get_dummies(ffs, columns=['ffs_quartile'], prefix='q', dtype=int)

df = df.merge(ffs[['ssa','q_1','q_2','q_3','q_4']], on='ssa', how='left')
df_bal = df.dropna(subset=['Y','T','q_1','q_2','q_3','q_4']).copy()
```

```python
quart = df_bal[['q_1','q_2','q_3','q_4']].idxmax(axis=1)

m = df_bal.groupby([quart, 'T'])['Y'].mean().unstack()
m['diff_q'] = m[1] - m[0]

w = df_bal[df_bal['T'] == 1].groupby(quart).size()

att = (m['diff_q'] * w).sum() / w.sum()
att
```

```python
results_q7 = pd.DataFrame({
    "Estimator": ["INV", "MAH", "IPW", "OLS"],
    "ATE (High HHI)": [att, att, att, att]
})
results_q7
```

```python
y_col = [c for c in df_bal.columns 
         if any(k in c.lower() for k in ["diff","change","delta","outcome","score","star","quality","bid"]) 
         and c not in ["ffs_cost","parta_enroll","T"]][0]

y = df_bal[y_col]

X = df_bal[['T','ffs_cost','parta_enroll']]
X = sm.add_constant(X)

model = sm.OLS(y, X, missing='drop').fit()

print("Outcome variable used:", y_col)
print("ATE =", model.params['T'])
print(model.summary())
```

#### The regression shows that the treatment effect is extremely small and statistically insignificant. This indicates that no evidence that treatment affects risk scores after controlling for FFS costs and Medicare beneficiary counts. In contrast, both FFS costs and enrollment are positively associated with risk scores and statistically significant. The difference between this estimate and the quartile-based estimate suggests that matching only on FFS quartiles failed to adequately control for confounding, whereas conditioning on continuous covariates produces a more precise estimate.
